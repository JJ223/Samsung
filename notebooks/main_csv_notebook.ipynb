```python
import pandas as pd
import numpy as np
import os
import re
```

```python
path = os.path.join('..', 'data', 'budget-data-new')

files = os.listdir(path)

if '.DS_Store' in files:
    files.remove('.DS_Store')

files[:5]
```

```python

def extract_company_name_and_quote_id(file_name):
    split = file_name.split('_')
    return split[0], split[1].strip('.tsv')

new_df_rows = []

for file in files:
    df = pd.read_csv(os.path.join(path, file), sep='\t', encoding='utf-8')
    
    company_name, quote_id = extract_company_name_and_quote_id(file)
    
    for index, row in df.iterrows():
        new_df_rows.append([
            company_name,
            quote_id,
            row['dimensions'],
            row['qtd'],
            row['punit']
        ])
    
```

```python
new_df = pd.DataFrame(
    columns=['company_name', 'quote_id', 'description', 'qty', 'unit_price'],
    data=new_df_rows
)

new_df
```

```python
new_df.to_csv('../data/budget-data.csv', index=False, encoding='utf-8')
```
